````python
import pandas as pd
fonte = "https://github.com/daniel-usp/LIT/blob/master/Enem_2019.csv?raw=true"
dados = pd.read_csv(fonte)
print(dados)
````

````python
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Sem o index o value_counts exibe os valores quem mais aparecem, por exemplo: Com 18 anos prestaram Enem 21452 pessoas, com 16 anos prestaram Enem 21256 pessoas...
dados["NU_IDADE"].value_counts()
````

````python
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Com 13 anos prestramenem 4, com 14 anos orestaram 141 pessoas...
dados["NU_IDADE"].value_counts().sort_index()
````

````python
#Plotar Histograma
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Plote de 13 até 40 anos
dados["NU_IDADE"].value_counts().plot.bar()
````

````python
#Plotar Histograma
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Plote de 13 até 40 anos. Graças ao comando loc é possível segregar um grupo desejado
dados.loc[(dados["NU_IDADE"] >=13) & (dados["NU_IDADE"] <= 40), "NU_IDADE"].value_counts().plot.bar()
````

````python
#Uso da query - Segregação sem o uso do loc
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Plote de 13 até 40 anos. Graças ao comando loc é possível segregar um grupo desejado
dados.query("NU_IDADE >= 13 and NU_IDADE <= 40")["NU_IDADE"]
````

````python
#Uso da query - Segregação sem o uso do loc (OBSRVAR QUE O COMANDO QUERY EXIGE ABRIR E FECHAR COM ASPAS SIMPLES)
#QUERY = CONSULTA ('filtros')
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Plote de 13 até 40 anos. Graças ao comando loc é possível segregar um grupo desejado
dados.query("NU_IDADE >= 13 and NU_IDADE <= 40")["NU_IDADE"].value_counts().plot.bar()
````

````python
#Plotar a quantidade de alunos por idade (NU_IDADE)
#Plote de 13 até 40 anos.
idade_13_40_anos = dados.query("NU_IDADE >= 13 and NU_IDADE <= 40 and TP_SEXO == 'M'")["NU_IDADE"]
idade_13_40_anos.value_counts().sort_index().plot.bar()
````

````python
#Plotar as notas de redação dos homens e das mulheres (NU_NOTA_REDACAO)
#query = consulta ('filtros')
idade_das_alunas = dados.query('TP_SEXO == "F"')['NU_IDADE']
idade_das_alunas.hist(bins=20)

idade_dos_alunos = dados.query('TP_SEXO == "M"')['NU_IDADE']
idade_dos_alunos.hist(bins=20)

#coloque as legendas no gráfico



````

````python
 ### Código: ```python import pandas as pd import matplotlib.pyplot as plt
 # Supondo que 'dados' seja um DataFrame do pandas já carregado com as colunas podem
 # Filtros para as idades
 import matplotlib.pyplot as plt
 redacao_das_alunas = dados.query('TP_SEXO == "F"')['NU_NOTA_REDACAO']
 redacao_dos_alunos = dados.query('TP_SEXO == "M"')['NU_NOTA_REDACAO']
 # Configuração do gráfico plt.figure(figsize=(10, 6))
 # Define o tamanho do gráfico
 # Histogramas para as redações
 plt.hist (redacao_das_alunas, bins=20, alpha=0.7, label='Mulheres', color='orange')
 plt.hist(redacao_dos_alunos, bins=20, alpha=0.7, label='Homens', color='blue')
 # Adicionado título e rótulos aos eixos
 plt.title('Distribuição de Redação por Sexo', fontsize=14)
 plt.xlabel('Redacao', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
 # Adiciona legenda
 plt.legend()
 # Mostra o gráfico
 plt.show()





````
